### seqCleaner

```python
#!/usr/bin/env python3 
# Name: Marco Lanza (mnlanza)
# Group Members: Adrian Flores (aflor115), Noah Kirsch (nkirsch), Chris

'''
Read a DNA string from user input and return a collapsed substring of embedded Ns to: {count}.

Example: 
 input: AaNNNNNNGTC
output: AA{6}GTC

Any lower case letters are converted to uppercase
'''
import re # imports python module used in purify

class DNAstring (str):
  def length (self):
    return (length(self))
  def purify(self):
    ''' Return an upcased version of the string, collapsing a single run of Ns.'''
    # Converts to the original string to uppercase and replace runs of N's with their {count}
    return re.sub(r'N+', lambda m: f'{{{len(m.group(0))}}}', self.upper()) 
    #re.sub Replaces occurrences of the pattern with a specified replacement string.
    #'N+' finds all of the occurences of N
    #lamda defines a new function and m is a match object that is passed to the functioned by the re.sub() method
    '''The `group(0)` method of the match object `m` returns the entire substring that was matched by the regex pattern. 
    For example, if the pattern matched "NNN", then `m.group(0)` would return "NNN'''
    #so len m.group tells us how many N's are in the group
    #The `{{` and `}}` are used to include literal curly braces in the output string

      
def main():
  ''' Get user DNA data and clean it up.'''
  data = input('DNA data?')
  while (data):
    thisDNA = DNAstring (data)
    pureData = thisDNA.purify()
    print (pureData)
    data = input('DNA data?')
    
main()
```

### fastqParse

```python
#!/usr/bin/env python3 
# Name: Marco Lanza (mnlanza)
# Group Members: Adrian Flores (aflor115), Noah Kirsch (nkirsch), Chris

'''
This program takes a FASTQ seqname as an input and outputs the various meanings/values held by the seqname
'''

class FastqString (str):
    def length (self):
        return (len(self))
    ''' class FastqString is to hold the length() function and be able to parse a FastqString using parse()'''
    def parse(self):
    #Goes through the FASTQ seqname and sepparates the seqname into its attributes
        components = self.strip().split(':') #strips leading and trailing whitespace and splits the components of seqname into a list

        # Assign each component to a variable
        return { #assigns each substring to a key in  a dictionary
            "Instrument": components[0],
            "Run ID": components[1],
            "Flow Cell ID": components[2],
            "Flow Cell Lane": components[3],
            "Tile Number": components[4],
            "X-coord": components[5],
            "Y-coord": components[6]
        }
    
def main():
    data = input('Input the name of the seqname line of a FASTQ: ')
    fastq_string = FastqString(data)  # Create an instance of FastqString
    parsed_data = fastq_string.parse()  # Parse the data

    for key, value in parsed_data.items(): #prints all of the keys and values in the dictionary 
        print(f"{key} = {value}")


main()
```

### Hints: 
 - Each coordinate will contain only 3 numeric values. 
 - Bond lengths are the distance between two points, which for points P and Q in 3-space,
$(P_x , P_y , P_z )$ and $(Q_x , Q_y , Q_z )$ respectively, the distance between them is:


$$
\begin {eqnarray}
     \|PQ\| &=&\sqrt{(P_x-Q_x)^2 + (P_y-Q_y)^2 + (P_z-Q_z)^2)} \\
            &=&\sqrt{\sum^{i \in x,y,z}{(P_i-Q_i)^2}} \\
\end{eqnarray}
$$

 - Bond angles can be calculated from the dot product.
 
Let’s say that we have three points in space labeled P, Q and R. We are interested in the angle at point Q that is made from line segments QP and QR.
The dot product tells us (for standard vectors P and R) that:
$$P \cdot R=\|P\| \|R\| \cos{\theta}$$
in this notation,$\|P\|$ refers to the length of vector P as a standard vector (assumed to begin at the origin (0,0,0) ).
We can then see that the angle between vectors P and R can be found by:
$$\cos{\theta}=\frac{P \cdot R} {\|P\| \|R\|}$$

We can calculate the dot product using the sum of products of the vector components:
$$P \cdot R = \sum^{i \in x,y,z}{P_i R_i} $$

Now, to find vector P in standard form, we need to remember that QP starts at Q, so we need to place the origin at Q and find out where P is in that new space. We do that by subtracting the components of Q from P.
Putting all of this together, we get:
$$\theta=\cos^{-1}{\frac{\sum^{i \in x,y,z}{(P_i - Q_i) (R_i - Q_i) }} {\|QP\| \|QR\|}}$$
Remember, $\theta$ is in radians.

### coordinateMathSoln

```python
#!/usr/bin/env python3 
# Name: Marco Lanza (mnlanza)
# Group Members: Adrian Flores (aflor115), Noah Kirsch (nkirsch), Chris

'''
This program takes the bond angles between three coordinates and calculates the bond angles and lengths 
'''

import math


class Triad :
    """
    Calculate angles and distances among a triad of points.
 
    Author: David Bernick
    Date: March 21, 2013
    Points can be supplied in any dimensional space as long as they are consistent.
    Points are supplied as tupels in n-dimensions, and there should be three
    of those to make the triad. Each point is positionally named as p,q,r
    and the corresponding angles are then angleP, angleQ and angleR.
    Distances are given by dPQ(), dPR() and dQR()
 
    Required Modules: math
    initialized: 3 positional tuples representing Points in n-space
             p1 = Triad( p=(1,0,0), q=(0,0,0), r=(0,1,0) )
    attributes: p,q,r the 3 tuples representing points in N-space
    methods:  angleP(), angleR(), angleQ() angles measured in radians
          dPQ(), dPR(), dQR() distances in the same units of p,q,r
 
    """

 
    def __init__(self,p,q,r) :
        """ Construct a Triad. 
        
        Example construction:
            p1 = Triad( p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.) ). 
        """

        self.p = p
        self.q = q
        self.r = r
        
# private helper methods
    def d2 (self,a,b) : # calculate squared distance of point a to b
        return float(sum((ia-ib)*(ia-ib)  for  ia,ib in zip (a,b)))
    
    def dot (self,a,b) : # dotProd of standard vectors a,b
        return float(sum(ia*ib for ia,ib in zip(a,b)))
    
    def ndot (self,a,b,c) : # dotProd of vec. a,c standardized to b
        return float(sum((ia-ib)*(ic-ib) for ia,ib,ic in zip (a,b,c)))
    
# calculate lengths(distances) of segments PQ, PR and QR
    def dPQ (self):
        """ Provides the distance between point p and point q """
        return math.sqrt(self.d2(self.p,self.q))
    
    def dPR (self):
        """ Provides the distance between point p and point r """
        return math.sqrt(self.d2(self.p,self.r))
    
    def dQR (self):
        """ Provides the distance between point q and point r """
        return math.sqrt(self.d2(self.q,self.r))
    
    def angleP (self) :
        """ Provides the angle made at point p by segments pq and pr (radians). """
        return math.acos(self.ndot(self.q,self.p,self.r) / 
                         math.sqrt(self.d2(self.q,self.p) * self.d2(self.r,self.p)))
    
    def angleQ (self) :
        """ Provides the angle made at point q by segments qp and qr (radians). """
        return math.acos(self.ndot(self.p,self.q,self.r) /
                         math.sqrt(self.d2(self.p,self.q) * self.d2(self.r,self.q)))
 
    def angleR (self) :
        """ Provides the angle made at point r by segments rp and rq (radians). """
        return math.acos(self.ndot(self.p,self.r,self.q) /
                         math.sqrt(self.d2(self.p,self.r) * self.d2(self.q,self.r)))

class Triad:
    """
    Calculate angles and distances among a triad of points.

    Author: David Bernick
    Date: March 21, 2013

    Points can be supplied as tuples in n-dimensions, and there should be three
    of those to make the triad. Each point is positionally named as p, q, r
    and the corresponding angles are then angleP, angleQ and angleR.
    Distances are given by dPQ(), dPR(), dQR().

    Required Modules: math
    """

    def __init__(self, p, q, r):
        """Construct a Triad.
           Example: Triad(p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.))
        """
        self.p = p  # We'll treat as C
        self.q = q  # We'll treat as N
        self.r = r  # We'll treat as Ca

    # --- Private helper methods ---
    def d2(self, a, b):
        """Calculate squared distance of point a to b."""
        return float(sum((ia - ib) ** 2 for ia, ib in zip(a, b)))

    def ndot(self, a, b, c):
        """Dot product of vectors a->b and c->b."""
        return float(sum((ia - ib) * (ic - ib) for ia, ib, ic in zip(a, b, c)))

    # --- Distance calculations ---
    def dPQ(self):
        """Distance between point p and point q."""
        return math.sqrt(self.d2(self.p, self.q))

    def dPR(self):
        """Distance between point p and point r."""
        return math.sqrt(self.d2(self.p, self.r))

    def dQR(self):
        """Distance between point q and point r."""
        return math.sqrt(self.d2(self.q, self.r))

    # --- Angle calculations (in radians) ---
    def angleP(self):
        """Angle at point p (C) by segments p->q and p->r."""
        return math.acos(
            self.ndot(self.q, self.p, self.r)
            / math.sqrt(self.d2(self.q, self.p) * self.d2(self.r, self.p))
        )

    def angleQ(self):
        """Angle at point q (N) by segments q->p and q->r."""
        return math.acos(
            self.ndot(self.p, self.q, self.r)
            / math.sqrt(self.d2(self.p, self.q) * self.d2(self.r, self.q))
        )

    def angleR(self):
        """Angle at point r (Ca) by segments r->p and r->q."""
        return math.acos(
            self.ndot(self.p, self.r, self.q)
            / math.sqrt(self.d2(self.p, self.r) * self.d2(self.q, self.r))
        )

def main():
    """
    Accepts a single line of input in the format:
      C = (x1, y1, z1) N = (x2, y2, z2) Ca = (x3, y3, z3)
    Then prints out:
      N-C bond length
      N-Ca bond length
      C-N-Ca bond angle
    """

    # Example input line:
    # C = (39.447, 94.657, 11.824) N = (39.292, 95.716, 11.027) Ca = (39.462, 97.101, 11.465)
    line = input("Enter coordinates (one line): ")

    # 1) Split the input by '(' to isolate each coordinate part.
    #    Example result: ["C = ", "39.447, 94.657, 11.824) N = ", "39.292, 95.716, 11.027) Ca = ", "39.462, 97.101, 11.465)"]
    parts = line.split("(")

    # We expect 4 parts: 0 is "C = ", 1 is "39.447, 94.657, 11.824) N = "
    #                    2 is "39.292, 95.716, 11.027) Ca = ", 3 is "39.462, 97.101, 11.465)"
    if len(parts) != 4:
        print("Invalid input format. Please try again.")
        return

    # 2) For each coordinate portion, split by ')' and take the first element
    #    to remove trailing text.
    c_str = parts[1].split(")")[0]
    n_str = parts[2].split(")")[0]
    ca_str = parts[3].split(")")[0]

    # 3) Convert the coordinate strings into tuples of floats
    #    e.g. "39.447, 94.657, 11.824" -> (39.447, 94.657, 11.824)
    def to_float_tuple(s):
        return tuple(map(float, s.split(',')))

    c_coords = to_float_tuple(c_str)
    n_coords = to_float_tuple(n_str)
    ca_coords = to_float_tuple(ca_str)

    # Create a Triad instance
    triad = Triad(p=c_coords, q=n_coords, r=ca_coords)

    # Distances
    dCN   = triad.dPQ()  # distance between C (p) and N (q)
    dNCa  = triad.dQR()  # distance between N (q) and Ca (r)

    # Angle at N => "C-N-Ca"
    angleN_rad = triad.angleQ()
    angleN_deg = angleN_rad * 180.0 / math.pi

    # Print results in the desired format
    print(f"N-C bond length = {dCN:.2f}")
    print(f"N-Ca bond length = {dNCa:.2f}")
    print(f"C-N-Ca bond angle = {angleN_deg:.1f}")

if __name__ == "__main__":
    main()
```
